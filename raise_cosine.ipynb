```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.solve_ivp.html#scipy.integrate.solve_ivp
from scipy.integrate import *
import numpy as np
from matplotlib.pyplot import *
from cmath import *
from scipy.signal import *
from scipy import signal
```

```python
def sinc(t):
    if t==0:
        return 1
    else:
        return np.sin(np.pi*t)/(np.pi*t)
def rcos(t,beta,T):
    a = np.zeros(len(t))
    for i in range(len(t)):
        if not (beta==0): 
            if (t[i]==T/2/beta or t[i]==-T/2/beta):
                a[i] = np.pi/4/T*( sinc(1/2/beta))
        
        a[i]=1/T*sinc(t[i]/T)*np.cos(np.pi*beta*t[i]/T)/(1-(2*beta*t[i]/T)**2)
    return a
```

```python
t0=1e-9
t = np.arange(-0.3,3,0.001)*t0
T = 0.1e-9
beta  = 1
# sinc
figure()
plot(t, rcos(t,0,T))
plot(t,np.zeros(len(t)))
# plt.show()
# raise consine

# plt.figure()
plot(t, rcos(t,beta,T))
plot(t,np.zeros(len(t)))
show()
```
